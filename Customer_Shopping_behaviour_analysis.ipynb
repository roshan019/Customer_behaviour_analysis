```python
# look is there any null value or not and sum all the no. of null on column

df.isnull().sum()
```

```python
#- groupby('Category'): Splits the DataFrame into groups based on the Category column.
#'Review Rating']: Focuses only on the Review Rating column within each group.
#.transform(...): Ensures the result aligns back with the original DataFrame’s index (so you don’t lose structure).
#lambda x: x.fillna(x.median()): For each group, replaces missing values (NaN) in Review Rating with the median of that group.
#Why this is useful

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
#- In snake case, all words are written in lowercase.
#Spaces are replaced with underscores (_).


#- Converts all column names in the DataFrame to lowercase.
df.columns = df.columns.str.lower()

#Replaces spaces in column names with underscores.
df.columns = df.columns.str.replace(' ', '_')

#- Renames a specific column: "purchase_amount_(usd)" becomes "purchase_amount"
df = df.rename(columns = {'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

Data loading and Exploratory data analysis (EDA)

```python
# Create a columns age_group

#- Creates a list of category names (labels) that will be used to label age groups.
#- These are custom labels you’ve chosen to represent four age ranges.
labels = ['young adult','adult','middle_aged','senior']

#- Uses pd.qcut to divide the age column in your DataFrame (df) into 4 quantile-based bins (quartiles). - q=4 means the ages will be split into 4 groups, each containing roughly the same number of observations.
#- The labels=labels argument assigns the custom names (young adult, adult, middle_aged, senior) to each bin instead of numeric codes.
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age', 'age_group']].head(10)
```

```python
# Create columns purchase_frequency_days

frequency_mapping = {
    'Fortnighty' : 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
